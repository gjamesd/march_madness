```python
def get_data(string):
    lst = []
    
    split_list = string.split('<circle')
    for i in range(len(split_list)):
        lst.append(split_list[i])
    
    
    temp = pd.DataFrame(columns = list(range(0, 9)))
    
    for i in range(1, len(split_list)):
        temp = temp.append(pd.Series(split_list[i].split(" ")),  ignore_index=True)
    
    temp = temp.drop(columns = [0, 8, 9])
    temp = temp.rename(columns = {1:"cx", 2:'cy', 
                       3:'r', 4:'fill', 
                       5:'stroke', 6:'opacity', 7:'style' })
     
    return temp
```

```python

def clean_data(df):
    data['cx'] = data['cx'].apply(lambda x: float(x.replace('cx="', '').replace('"', '')))
    data['cy'] = data['cy'].apply(lambda x: float(x.replace('cy="', '').replace('"', '')))
    data['r'] = data['r'].apply(lambda x: float(x.replace('r="', '').replace('"', '')))
    data['fill'] = data['fill'].apply(lambda x: x.replace('fill="', '').replace('"', ''))
    data['stroke'] = data['stroke'].apply(lambda x: x.replace('stroke="', '').replace('"', ''))
    data['opacity'] = data['opacity'].apply(lambda x: x.replace('opacity="', '').replace('"', ''))
    data['opacity'] = data['opacity'].apply(lambda x: float(x.replace('opacity="', '').replace('"', '')))
    data['style'] = data['style'].apply(lambda x: x.replace('style="', '').replace(':', ''))
    data['result'] = data['fill'].apply(lambda x: "Make" if x == "#c80000" else "Miss")
    
    
    return data
```

```python
def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))*69
```

```python
def convert_to_meters(df):
    
    #this is the number of meters from the try line the kick is taken at
    df['x_meters'] = df['cx'].apply(lambda x: x/9.5)
    #converte to y meters with 0 being the top of the image, and 70 being the bottom of the image
    df['y_meters'] = NormalizeData(df['cy'])
    
    df['outcome'] = df['result'].apply(lambda x: 1 if x == "Make" else 0)
    
    return df 
```
